# Week 07 Matplotlib 히스토그램과 Seaborn 실습

이번 실습은 `정렬 -> 요약 통계 -> matplotlib 히스토그램 -> seaborn histplot -> lineplot -> scatterplot` 흐름으로 진행합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## 0. 샘플 데이터 만들기

```python
np.random.seed(42)

scores = np.random.normal(loc=70, scale=12, size=200)
group_a = np.random.normal(loc=65, scale=10, size=200)
group_b = np.random.normal(loc=75, scale=14, size=200)

df_scores = pd.DataFrame({
    'score': scores,
    'group': np.where(scores >= np.mean(scores), 'high_side', 'low_side')
})

df_scores.head()
```

## 1. `np.sort()`로 작은 값부터 큰 값까지 정렬하기

```python
sorted_scores = np.sort(scores)

print('앞 10개:', np.round(sorted_scores[:10], 2))
print('뒤 10개:', np.round(sorted_scores[-10:], 2))
```

## 2. 범위, 중심, 퍼짐을 먼저 보기

```python
score_min = np.min(scores)
score_max = np.max(scores)
score_mean = np.mean(scores)
score_median = np.median(scores)
score_std = np.std(scores)
q1, q2, q3 = np.percentile(scores, [25, 50, 75])

print(f'min: {score_min:.2f}')
print(f'max: {score_max:.2f}')
print(f'mean: {score_mean:.2f}')
print(f'median: {score_median:.2f}')
print(f'std: {score_std:.2f}')
print(f'Q1: {q1:.2f}, Q2: {q2:.2f}, Q3: {q3:.2f}')
```

## 3. 가장 기본적인 `matplotlib` 히스토그램

```python
plt.figure(figsize=(7, 4))
plt.hist(scores)
plt.title('Basic Histogram')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

## 4. `bins`를 직접 지정하기

```python
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=12, color='skyblue', edgecolor='black')
plt.title('Histogram with bins=12')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

## 5. `bins` 개수 차이 비교

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, bin_count in zip(axes, [5, 12, 30]):
    ax.hist(scores, bins=bin_count, color='lightcoral', edgecolor='black')
    ax.set_title(f'bins={bin_count}')
    ax.set_xlabel('score')
    ax.set_ylabel('count')

plt.tight_layout()
plt.show()
```

## 6. 색상, 경계선, 투명도

```python
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=15, color='seagreen', edgecolor='white', alpha=0.8)
plt.title('Color, Edge, Alpha')
plt.xlabel('score')
plt.ylabel('count')
plt.grid(axis='y', alpha=0.25)
plt.show()
```

## 7. 두 분포를 겹쳐서 비교하기

```python
plt.figure(figsize=(8, 4))
plt.hist(group_a, bins=15, alpha=0.5, label='group_a', color='royalblue')
plt.hist(group_b, bins=15, alpha=0.5, label='group_b', color='orange')
plt.title('Overlayed Histograms')
plt.xlabel('score')
plt.ylabel('count')
plt.legend()
plt.show()
```

## 8. `density=True`로 보기

```python
plt.figure(figsize=(8, 4))
plt.hist(group_a, bins=15, alpha=0.5, density=True, label='group_a', color='royalblue')
plt.hist(group_b, bins=15, alpha=0.5, density=True, label='group_b', color='orange')
plt.title('Density Histogram')
plt.xlabel('score')
plt.ylabel('density')
plt.legend()
plt.show()
```

## 9. `cumulative=True`로 누적 히스토그램

```python
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=20, cumulative=True, color='mediumpurple', edgecolor='black')
plt.title('Cumulative Histogram')
plt.xlabel('score')
plt.ylabel('cumulative count')
plt.show()
```

## 10. `ax.hist()`로 축 객체 기준으로 그리기

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(scores, bins=15, color='steelblue', edgecolor='black')
ax.set_title('Histogram drawn with ax.hist()')
ax.set_xlabel('score')
ax.set_ylabel('count')
plt.show()
```

## 11. `Pandas` 열과 `sort_values()`

```python
sorted_series = df_scores['score'].sort_values()

print('오름차순 앞 5개')
print(np.round(sorted_series.head(5), 2))
print('\n오름차순 뒤 5개')
print(np.round(sorted_series.tail(5), 2))
```

## 12. `Pandas` 열을 바로 히스토그램에 넣기

```python
plt.figure(figsize=(7, 4))
plt.hist(df_scores['score'], bins=15, color='tomato', edgecolor='black')
plt.title('Histogram from a Pandas Series')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

## 13. `seaborn` 기본 스타일 적용

```python
sns.set_theme(style='whitegrid', palette='deep')
```

## 14. 산점도용 학습 데이터 만들기

```python
study_hours = np.random.uniform(1, 8, size=60)
sleep_hours = np.random.uniform(4.5, 8.5, size=60)
group = np.random.choice(['A', 'B'], size=60)
study_score = 45 + study_hours * 6 + sleep_hours * 2 + np.where(group == 'A', 4, -2) + np.random.normal(0, 5, size=60)
passed = np.where(study_score >= 75, 'pass', 'fail')

df_study = pd.DataFrame({
    'study_hours': study_hours,
    'sleep_hours': sleep_hours,
    'score': study_score,
    'group': group,
    'passed': passed
})

df_study.head()
```

## 15. `fmri` 데이터 불러오기

```python
fmri = pd.read_csv('data/fmri.csv')
fmri.head()
```

## 16. `sns.histplot()` 기본형

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df_scores, x='score', bins=15)
plt.title('Basic seaborn histplot')
plt.show()
```

## 17. `kde=True`로 분포 곡선 같이 보기

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df_scores, x='score', bins=15, kde=True, color='teal')
plt.title('histplot with KDE')
plt.show()
```

## 18. `hue`와 `stat='density'`로 그룹 비교하기

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df_scores, x='score', hue='group', bins=15, stat='density', alpha=0.45)
plt.title('Grouped density histplot')
plt.show()
```

## 19. `fmri` 데이터로 가장 기본적인 선그래프 그리기

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=fmri, x='timepoint', y='signal')
plt.title('Basic lineplot on fmri dataset')
plt.show()
```

## 20. `hue`와 `style`을 함께 써서 `fmri` 선그래프 비교하기

```python
plt.figure(figsize=(9, 5))
sns.lineplot(
    data=fmri,
    x='timepoint',
    y='signal',
    hue='event',
    style='region'
)
plt.title('fmri signal by event and region')
plt.show()
```

## 21. 가장 기본적인 `sns.scatterplot()`

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df_study, x='study_hours', y='score')
plt.title('Basic scatterplot')
plt.show()
```

## 22. `hue`와 `style`로 그룹을 함께 표현하기

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_study, x='study_hours', y='score', hue='group', style='passed')
plt.title('Scatterplot with hue and style')
plt.show()
```

## 23. `size`로 추가 수치 정보 넣기

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df_study,
    x='study_hours',
    y='score',
    hue='group',
    size='sleep_hours',
    sizes=(30, 220),
    alpha=0.75
)
plt.title('Scatterplot with hue and size')
plt.show()
```

## 24. `matplotlib`의 `ax`와 함께 쓰기

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(data=df_study, x='study_hours', y='score', hue='group', ax=ax)
ax.set_title('Seaborn scatterplot on Matplotlib Axes')
ax.set_xlabel('study_hours')
ax.set_ylabel('score')
plt.show()
```

## 25. `FacetGrid`로 그룹별 산점도 나누어 보기

```python
g = sns.FacetGrid(df_study, col='group', hue='passed', height=4)
g.map_dataframe(sns.scatterplot, x='study_hours', y='score')
g.add_legend()
g.set_axis_labels('study_hours', 'score')
g.set_titles(col_template='group = {col_name}')
```

## 26. `FacetGrid`로 `fmri` 선그래프 나누어 보기

```python
g = sns.FacetGrid(fmri, row='region', col='event', height=3, margin_titles=True)
g.map_dataframe(sns.lineplot, x='timepoint', y='signal')
g.set_axis_labels('timepoint', 'signal')
```

## 27. `relplot()`으로 scatter + 패널을 더 쉽게 그리기

```python
g = sns.relplot(
    data=df_study,
    x='study_hours',
    y='score',
    hue='passed',
    col='group',
    kind='scatter',
    height=4,
    aspect=1
)
g.fig.suptitle('relplot: scatter panels by group', y=1.03)
```

## 28. `displot()`으로 분포를 패널로 나누기

```python
g = sns.displot(
    data=df_scores,
    x='score',
    col='group',
    kind='hist',
    bins=12,
    height=4,
    aspect=1
)
g.fig.suptitle('displot: histogram panels by group', y=1.03)
```

## 29. `catplot()`과 `pairplot()` 맛보기

```python
g = sns.catplot(data=df_study, x='group', hue='passed', kind='count', height=4, aspect=1.2)
g.fig.suptitle('catplot: count by group and passed', y=1.03)

sns.pairplot(df_study[['study_hours', 'sleep_hours', 'score', 'group']], hue='group')
```

## 30. `Plotly`의 `iris` 데이터셋 불러오기

```python
iris = px.data.iris()
iris.head()
```

## 31. `px.scatter()`로 interactive 관계 보기

```python
fig = px.scatter(
    iris,
    x='sepal_width',
    y='sepal_length',
    color='species',
    size='petal_length',
    hover_data=['petal_width']
)
fig.show()
```

## 32. `px.histogram()`과 `px.box()`로 분포 보기

```python
fig = px.histogram(iris, x='sepal_length', color='species', barmode='overlay')
fig.show()

fig = px.box(iris, x='species', y='petal_length', color='species')
fig.show()
```

## 33. `px.line()`으로 `fmri` 집계 추세 보기

```python
fmri_mean = (
    fmri.groupby(['timepoint', 'event'], as_index=False)['signal']
    .mean()
)

fig = px.line(
    fmri_mean,
    x='timepoint',
    y='signal',
    color='event',
    markers=True
)
fig.show()
```
