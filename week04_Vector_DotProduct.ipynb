## 0. 자주 쓰는 배열 연산 함수

내적을 보기 전에, NumPy에서 자주 쓰는 배열 내부 연산 함수, 비교연산, `axis` 개념을 먼저 확인한다.
배열의 같은 위치 원소끼리 계산하는 방식과 불리언 비교 방식이 뒤 실습의 기반이 된다. 더 자세한 배열 기초는 3주차 노트로 연결한다.


```python
ops_matrix = np.array([[1, 2, 3], [4, 5, 6]])

print("ops_matrix =\n", ops_matrix)
print("np.sum(ops_matrix) =", np.sum(ops_matrix))
print("np.mean(ops_matrix) =", np.mean(ops_matrix))
print("np.max(ops_matrix) =", np.max(ops_matrix))
print("np.min(ops_matrix) =", np.min(ops_matrix))
```

설명: 위 함수들은 배열 전체를 한 번에 요약한다. `sum`은 합계, `mean`은 평균, `max`는 최댓값, `min`은 최솟값이다.

```python
print("행 기준 합계 axis=1 ->", np.sum(ops_matrix, axis=1))
print("열 기준 합계 axis=0 ->", np.sum(ops_matrix, axis=0))
print("행 기준 평균 axis=1 ->", np.mean(ops_matrix, axis=1))
print("열 기준 평균 axis=0 ->", np.mean(ops_matrix, axis=0))
```

설명: `axis=0`은 열 방향 계산, `axis=1`은 행 방향 계산이다. 2차원 배열에서 `sum`과 `mean`을 이해하려면 `axis` 개념이 중요하다.

## 0-1. 원소별 배열 연산

배열끼리 더하거나 곱할 때는 같은 위치의 원소끼리 계산한다. 이 흐름이 그대로 내적 계산의 출발점이 된다.


```python
arr_x = np.array([1, 2, 3])
arr_y = np.array([4, 5, 6])

print("arr_x =", arr_x)
print("arr_y =", arr_y)
print("arr_x + arr_y =", arr_x + arr_y)
print("arr_x * arr_y =", arr_x * arr_y)
```

설명: `arr_x + arr_y`는 같은 위치 원소를 더하고, `arr_x * arr_y`는 같은 위치 원소를 곱한다.
`dot`은 이렇게 곱한 결과를 다시 모두 더하는 연산으로 볼 수 있다.


## 0-2. 비교연산

배열을 어떤 값과 비교하면 각 원소마다 조건을 검사한 `True`, `False` 배열이 나온다.


```python
compare_arr = np.array([1, 3, 5, 7])

print("compare_arr =", compare_arr)
print("compare_arr > 3 =", compare_arr > 3)
print("compare_arr == 5 =", compare_arr == 5)
print("compare_arr % 2 == 1 =", compare_arr % 2 == 1)
```

설명: 비교연산 결과는 하나의 값이 아니라 불리언 배열이다.
NumPy는 여러 값을 한 번에 비교할 수 있고, 이 결과는 나중에 조건 필터링의 기준으로도 쓸 수 있다.


## 1. 내적 계산 기초

두 벡터의 같은 위치 원소를 곱한 뒤 모두 더하면 내적이 된다.

```python
b = np.array([1, 3])
c = np.array([4, 2])

print("b =", b)
print("c =", c)
print("b.dot(c) =", b.dot(c))
```

설명: `b.dot(c)`는 `1*4 + 3*2 = 10`이다. 즉 내적은 단순히 곱해서 더하는 계산이지만, 의미는 벡터 방향과 매우 깊게 연결된다.

## 2. 같은 방향일 때 내적

같은 방향을 보는 벡터는 내적이 양수이며, 보통 값도 크게 나온다.

```python
same_a = np.array([1, 2])
same_b = np.array([2, 4])

print("same_a =", same_a)
print("same_b =", same_b)
print("np.dot(same_a, same_b) =", np.dot(same_a, same_b))
```

설명: `same_b`는 `same_a`와 같은 방향이다. 그래서 내적은 양수이고, 두 벡터가 같은 방향으로 얼마나 함께 움직이는지 잘 보여 준다.

## 3. 수직일 때 내적

두 벡터가 서로 직각이면 내적은 0이 된다.

```python
perp_a = np.array([1, 0])
perp_b = np.array([0, 2])

print("perp_a =", perp_a)
print("perp_b =", perp_b)
print("np.dot(perp_a, perp_b) =", np.dot(perp_a, perp_b))
```

설명: 수직이라는 것은 한 벡터가 다른 벡터 방향으로 기여하는 정도가 없다는 뜻이다. 그래서 내적은 0이다.

## 4. 반대 방향일 때 내적

두 벡터가 반대 방향이면 내적은 음수가 된다.

```python
opp_a = np.array([1, 2])
opp_b = np.array([-2, -4])

print("opp_a =", opp_a)
print("opp_b =", opp_b)
print("np.dot(opp_a, opp_b) =", np.dot(opp_a, opp_b))
```

설명: 반대 방향은 서로 밀어내거나 충돌하는 관계로 볼 수 있다. 그래서 내적은 0보다 작다.

## 5. 사람 비유로 이해하기

내적은 두 사람이 같은 목표 방향으로 협력하는지, 서로 무관한지, 반대로 충돌하는지를 수치로 본다고 생각하면 이해하기 쉽다.

```python
person_same_1 = np.array([3, 1])
person_same_2 = np.array([6, 2])
person_perp_1 = np.array([1, 0])
person_perp_2 = np.array([0, 3])
person_opp_1 = np.array([2, 1])
person_opp_2 = np.array([-4, -2])

print("같은 방향 내적 =", np.dot(person_same_1, person_same_2))
print("수직 관계 내적 =", np.dot(person_perp_1, person_perp_2))
print("반대 방향 내적 =", np.dot(person_opp_1, person_opp_2))
```

설명: 같은 목표로 협력하면 양수, 서로 관계가 없으면 0, 서로 반대로 움직이면 음수라고 해석할 수 있다.

## 6. NumPy로 내적 계산하는 여러 방법

NumPy에서는 `.dot()`, `np.dot()`, `@` 연산자를 사용할 수 있다.

```python
a = np.array([1, 2, 3])
d = np.array([4, 5, 6])

print("a.dot(d) =", a.dot(d))
print("np.dot(a, d) =", np.dot(a, d))
print("a @ d =", a @ d)
```

설명: 세 방법은 같은 내적 결과를 계산한다. 벡터 단계에서는 세 표현을 모두 익혀 두면 좋다.

## 7. 내적과 방향성 연결하기

벡터의 방향을 좀 더 분명히 보기 위해 norm을 함께 확인한다.

```python
print("||same_a|| =", np.linalg.norm(same_a))
print("||perp_a|| =", np.linalg.norm(perp_a))
print("||opp_a|| =", np.linalg.norm(opp_a))
print("||opp_b|| =", np.linalg.norm(opp_b))
```

설명: 내적은 단순 계산이 아니라 방향성과 크기의 결합이다. 같은 방향이면 양수, 수직이면 0, 반대 방향이면 음수라는 해석이 핵심이다.
