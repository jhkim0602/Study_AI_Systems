# Week 08 Seaborn 입문 실습

이번 실습은 `histplot`, `lineplot`, `scatterplot`을 연결해 `seaborn`의 핵심 그래프를 이해하는 흐름으로 구성합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 0. 기본 스타일 적용

```python
sns.set_theme(style='whitegrid', palette='deep')
```

## 1. 샘플 점수 데이터 만들기

```python
np.random.seed(7)

math_score = np.random.normal(loc=72, scale=10, size=120)
english_score = np.random.normal(loc=66, scale=12, size=120)

df_score = pd.DataFrame({
    'score': np.concatenate([math_score, english_score]),
    'subject': ['math'] * len(math_score) + ['english'] * len(english_score)
})

df_score.head()
```

## 2. 산점도용 학습 데이터 만들기

```python
study_hours = np.random.uniform(1, 8, size=60)
sleep_hours = np.random.uniform(4.5, 8.5, size=60)
group = np.random.choice(['A', 'B'], size=60)
score = 45 + study_hours * 6 + sleep_hours * 2 + np.where(group == 'A', 4, -2) + np.random.normal(0, 5, size=60)
passed = np.where(score >= 75, 'pass', 'fail')

df_study = pd.DataFrame({
    'study_hours': study_hours,
    'sleep_hours': sleep_hours,
    'score': score,
    'group': group,
    'passed': passed
})

df_study.head()
```

## 3. `fmri` 데이터셋 불러오기

```python
fmri = pd.read_csv('StudyNote/week08/data/fmri.csv')
fmri.head()
```

## 4. 가장 기본적인 `sns.histplot()`

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df_score, x='score', bins=15)
plt.title('Basic seaborn histplot')
plt.show()
```

## 5. `kde=True`로 분포 곡선 같이 보기

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df_score, x='score', bins=15, kde=True, color='teal')
plt.title('histplot with KDE')
plt.show()
```

## 6. `hue`로 그룹별 비교하기

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df_score, x='score', hue='subject', bins=15, alpha=0.5)
plt.title('Grouped histplot with hue')
plt.show()
```

## 7. `stat='density'`로 정규화해서 보기

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df_score, x='score', hue='subject', bins=15, stat='density', alpha=0.45)
plt.title('Density-based histplot')
plt.show()
```

## 8. `fmri` 데이터로 가장 기본적인 선그래프 그리기

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=fmri, x='timepoint', y='signal')
plt.title('Basic lineplot on fmri dataset')
plt.show()
```

## 9. `hue`와 `style`을 함께 써서 `fmri` 선그래프 비교하기

```python
plt.figure(figsize=(9, 5))
sns.lineplot(
    data=fmri,
    x='timepoint',
    y='signal',
    hue='event',
    style='region'
)
plt.title('fmri signal by event and region')
plt.show()
```

## 10. 가장 기본적인 `sns.scatterplot()`

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df_study, x='study_hours', y='score')
plt.title('Basic scatterplot')
plt.show()
```

## 11. `hue`와 `style`로 그룹을 함께 표현하기

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_study, x='study_hours', y='score', hue='group', style='passed')
plt.title('Scatterplot with hue and style')
plt.show()
```

## 12. `size`로 추가 수치 정보 넣기

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df_study,
    x='study_hours',
    y='score',
    hue='group',
    size='sleep_hours',
    sizes=(30, 220),
    alpha=0.75
)
plt.title('Scatterplot with hue and size')
plt.show()
```

## 13. `matplotlib`의 `ax`와 함께 쓰기

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(data=df_study, x='study_hours', y='score', hue='group', ax=ax)
ax.set_title('Seaborn scatterplot on Matplotlib Axes')
ax.set_xlabel('study_hours')
ax.set_ylabel('score')
plt.show()
```

## 14. 이어서 배우게 될 Seaborn 함수 미리 보기

```python
tips_like = pd.DataFrame({
    'hours': np.random.randint(1, 8, size=30),
    'score': np.random.normal(70, 10, size=30),
    'team': np.random.choice(['A', 'B', 'C'], size=30)
})

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
sns.boxplot(data=tips_like, x='team', y='score', ax=axes[0])
axes[0].set_title('boxplot')
sns.countplot(data=tips_like, x='team', ax=axes[1])
axes[1].set_title('countplot')
plt.tight_layout()
plt.show()
```
