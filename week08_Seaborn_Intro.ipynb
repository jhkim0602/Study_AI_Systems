# Week 08 Seaborn 입문 실습

이번 실습은 `NumPy`, `Pandas`, `matplotlib`, `seaborn`을 함께 호출하는 구조에서 시작해, `histplot`과 `lineplot`을 `fmri` 데이터셋으로 연결합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 0. 기본 스타일 적용

```python
sns.set_theme(style='whitegrid', palette='deep')
```

## 1. 샘플 점수 데이터 만들기

```python
np.random.seed(7)

math_score = np.random.normal(loc=72, scale=10, size=120)
english_score = np.random.normal(loc=66, scale=12, size=120)

df_score = pd.DataFrame({
    'score': np.concatenate([math_score, english_score]),
    'subject': ['math'] * len(math_score) + ['english'] * len(english_score)
})

df_score.head()
```

## 2. `seaborn` 내장 `fmri` 데이터셋 불러오기

```python
fmri = pd.read_csv('StudyNote/week08/data/fmri.csv')
fmri.head()
```

## 3. 가장 기본적인 `sns.histplot()`

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df_score, x='score', bins=15)
plt.title('Basic seaborn histplot')
plt.show()
```

## 4. `kde=True`로 분포 곡선 같이 보기

```python
plt.figure(figsize=(7, 4))
sns.histplot(data=df_score, x='score', bins=15, kde=True, color='teal')
plt.title('histplot with KDE')
plt.show()
```

## 5. `hue`로 그룹별 비교하기

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df_score, x='score', hue='subject', bins=15, alpha=0.5)
plt.title('Grouped histplot with hue')
plt.show()
```

## 6. `stat='density'`로 정규화해서 보기

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df_score, x='score', hue='subject', bins=15, stat='density', alpha=0.45)
plt.title('Density-based histplot')
plt.show()
```

## 7. `fmri` 데이터로 가장 기본적인 선그래프 그리기

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=fmri, x='timepoint', y='signal')
plt.title('Basic lineplot on fmri dataset')
plt.show()
```

## 8. `hue`와 `style`을 함께 써서 `fmri` 선그래프 비교하기

```python
plt.figure(figsize=(9, 5))
sns.lineplot(
    data=fmri,
    x='timepoint',
    y='signal',
    hue='event',
    style='region'
)
plt.title('fmri signal by event and region')
plt.show()
```

## 9. `matplotlib`의 `ax`와 함께 쓰기

```python
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', ax=ax)
ax.set_title('Seaborn lineplot on Matplotlib Axes')
ax.set_xlabel('timepoint')
ax.set_ylabel('signal')
plt.show()
```

## 10. 이어서 배우게 될 Seaborn 함수 미리 보기

```python
tips_like = pd.DataFrame({
    'hours': np.random.randint(1, 8, size=30),
    'score': np.random.normal(70, 10, size=30),
    'team': np.random.choice(['A', 'B', 'C'], size=30)
})

fig, axes = plt.subplots(1, 3, figsize=(14, 4))
sns.scatterplot(data=tips_like, x='hours', y='score', hue='team', ax=axes[0])
axes[0].set_title('scatterplot')
sns.boxplot(data=tips_like, x='team', y='score', ax=axes[1])
axes[1].set_title('boxplot')
sns.countplot(data=tips_like, x='team', ax=axes[2])
axes[2].set_title('countplot')
plt.tight_layout()
plt.show()
```
