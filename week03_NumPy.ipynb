# Week 03: NumPy 기초 실습

이번 실습은 NumPy 입문 단계에서 자주 배우는 순서대로 구성되어 있다. 셀을 위에서부터 순서대로 실행하면서 배열 생성, 정규분포 난수, 속성 확인, 인덱싱, 슬라이싱, 연산, 집계, reshape를 누적형으로 익힌다.

```python
import numpy as np
import sys
import time
```

## 1. `np.array()`로 배열 만들기

파이썬 리스트를 NumPy 배열로 바꾸는 가장 기본적인 방법이다.

```python
arr = np.array([1, 2, 3, 4, 5])

print("arr =", arr)
print("type(arr) =", type(arr))
```

## 2. 기본 배열 생성 메서드

`zeros`, `ones`, `empty`, `full`, `arange`, `uniform`은 NumPy 입문에서 자주 사용하는 생성 메서드들이다.

```python
zeros_arr = np.zeros((2, 3))
ones_arr = np.ones((2, 3))
empty_arr = np.empty((2, 3))
full_arr = np.full((2, 3), 7)
range_arr = np.arange(1, 7)

np.random.seed(42)
uniform_arr = np.random.uniform(0, 1, 5)

print("zeros_arr =\n", zeros_arr)
print("ones_arr =\n", ones_arr)
print("empty_arr =\n", empty_arr)
print("full_arr =\n", full_arr)
print("range_arr =", range_arr)
print("uniform_arr =", uniform_arr)
```

## 2-1. 정규분포 난수 생성

`np.random.normal()`은 평균과 표준편차를 기준으로 정규분포를 따르는 난수를 만든다. `uniform`과 비교해 보면 분포의 차이를 이해하기 쉽다.

```python
np.random.seed(42)
normal_arr = np.random.normal(loc=50, scale=10, size=5)
normal_arr_large = np.random.normal(loc=50, scale=10, size=1000)

print("normal_arr =", np.round(normal_arr, 2))
print("normal_arr_large 평균 =", round(np.mean(normal_arr_large), 2))
print("normal_arr_large 표준편차 =", round(np.std(normal_arr_large), 2))
```

## 3. 배열의 기본 속성 확인

`shape`, `ndim`, `size`, `dtype`는 배열을 이해할 때 가장 먼저 확인하는 속성이다.

```python
matrix = range_arr.reshape(2, 3)

print("arr =", arr)
print("arr.shape =", arr.shape)
print("arr.ndim =", arr.ndim)
print("arr.size =", arr.size)
print("arr.dtype =", arr.dtype)

print()
print("matrix =\n", matrix)
print("matrix.shape =", matrix.shape)
print("matrix.ndim =", matrix.ndim)
print("matrix.size =", matrix.size)
print("matrix.dtype =", matrix.dtype)
print("matrix.T =\n", matrix.T)
```

## 4. 1차원 배열 인덱싱과 슬라이싱

앞에서 만든 `arr`를 이용해 값 선택과 구간 선택을 실습한다.

```python
print("arr =", arr)
print("arr[0] =", arr[0])
print("arr[2] =", arr[2])
print("arr[-1] =", arr[-1])

print("arr[1:4] =", arr[1:4])
print("arr[:3] =", arr[:3])
print("arr[2:] =", arr[2:])
print("arr[::2] =", arr[::2])
```

## 5. 2차원 배열 인덱싱과 슬라이싱

이번에는 `matrix`를 이용해 행과 열을 선택한다.

```python
print("matrix =\n", matrix)
print("matrix[0, 0] =", matrix[0, 0])
print("matrix[1, 2] =", matrix[1, 2])
print("matrix[0, :] =", matrix[0, :])
print("matrix[:, 1] =", matrix[:, 1])
print("matrix[:, :2] =\n", matrix[:, :2])
```

## 6. 배열 연산과 브로드캐스팅

NumPy는 배열 전체에 연산을 한 번에 적용할 수 있다.

```python
arr_plus_ten = arr + 10
arr_times_two = arr * 2
arr_plus_uniform = arr + uniform_arr
ones_vector = np.ones(5)

print("arr + 10 =", arr_plus_ten)
print("arr * 2 =", arr_times_two)
print("arr + uniform_arr =", arr_plus_uniform)
print("arr + ones_vector =", arr + ones_vector)
```

## 7. 자주 쓰는 집계 함수

`sum`, `mean`, `max`, `min`은 배열 데이터를 요약할 때 자주 사용한다.

```python
print("arr 합계 =", np.sum(arr))
print("arr 평균 =", np.mean(arr))
print("arr 최댓값 =", np.max(arr))
print("arr 최솟값 =", np.min(arr))

print("matrix 행별 합계 =", np.sum(matrix, axis=1))
print("matrix 열별 합계 =", np.sum(matrix, axis=0))
print("normal_arr 평균 =", np.mean(normal_arr))
```

## 8. 배열 모양 바꾸기

`reshape()`는 원소 수를 유지한 채 배열의 구조만 바꾼다.

```python
reshaped_3x2 = range_arr.reshape(3, 2)
reshaped_1x6 = range_arr.reshape(1, 6)

print("range_arr =", range_arr)
print("reshaped_3x2 =\n", reshaped_3x2)
print("reshaped_1x6 =\n", reshaped_1x6)
```

## 9. 미니 실습: 점수 배열 다루기

지금까지 배운 내용을 간단한 점수 데이터에 적용해 본다.

```python
scores = np.array([[90, 80, 100], [70, 85, 95]])

print("scores =\n", scores)
print("첫 번째 학생 점수 =", scores[0, :])
print("두 번째 과목 점수 =", scores[:, 1])
print("학생별 총점 =", np.sum(scores, axis=1))
print("과목별 평균 =", np.mean(scores, axis=0))
```

## 10. 리스트와 NumPy 배열의 메모리 비교

같은 수치 데이터를 저장하더라도 리스트와 NumPy 배열은 메모리 사용 방식이 다르다.

```python
list_data = [1, 2, 3, 4, 5]
array_data = np.array([1, 2, 3, 4, 5])

print("List 메모리 크기:", sys.getsizeof(list_data) * len(list_data))
print("NumPy 배열 메모리 크기:", array_data.nbytes)
```

## 11. 리스트와 NumPy 배열의 연산 속도 비교

큰 데이터에서는 리스트보다 NumPy 배열 연산이 더 빠른 경우가 많다.

```python
size = 10**6

list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result_list = [x + y for x, y in zip(list1, list2)]
end = time.time()
print("리스트 연산 시간:", end - start)

arr1 = np.array(list1)
arr2 = np.array(list2)

start = time.time()
result_array = arr1 + arr2
end = time.time()
print("NumPy 연산 시간:", end - start)
```

## 12. 고급 연산 실습

NumPy 배열 객체에서는 평균, 합계, 표준편차, 내적 같은 연산을 바로 수행할 수 있다.

```python
advanced_arr = np.array([1, 2, 3, 4, 5])

print("advanced_arr =", advanced_arr)
print(np.mean(advanced_arr))  # 평균
print(np.sum(advanced_arr))   # 합계
print(np.std(advanced_arr))   # 표준편차
print(np.dot(advanced_arr, advanced_arr))  # 내적 (dot product)
```

## 13. 벡터 기초 실습

NumPy의 1차원 배열은 벡터처럼 다룰 수 있다. 아래에서는 코드 셀과 설명 셀을 번갈아 보면서 벡터 연산을 따라간다.

```python
b = np.array([1, 3])
c = np.array([4, 2])

print("b =", b)
print("c =", c)
print("b.dot(c) =", b.dot(c))
```

설명: `b.dot(c)`는 벡터 내적입니다. `[1, 3]`과 `[4, 2]`의 내적은 `1*4 + 3*2 = 10`입니다.

```python
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

print("vector_a =", vector_a)
print("vector_b =", vector_b)
print("벡터 덧셈 =", vector_a + vector_b)
print("벡터 뺄셈 =", vector_b - vector_a)
```

설명: 벡터 덧셈과 뺄셈은 같은 위치의 원소끼리 계산합니다.

```python
print("스칼라 곱 =", 3 * vector_a)
print("벡터 내적 =", np.dot(vector_a, vector_b))
print("vector_a 크기 =", np.linalg.norm(vector_a))
```

설명: 스칼라 곱은 모든 원소에 같은 수를 곱하는 것이고, `np.linalg.norm()`은 벡터의 크기(길이)를 구합니다.
