# Week 03: NumPy 기초 실습

이번 실습은 NumPy의 기초만 다룬다. 배열 만들기, `shape` 확인, 기본 인덱싱, 간단한 연산을 직접 실행해 보자.

```python
import numpy as np
```

## 1. NumPy 배열 만들기

파이썬 리스트를 `np.array()`로 바꾸면 NumPy 배열을 만들 수 있다.

```python
arr = np.array([1, 2, 3, 4, 5])

print("배열:", arr)
print("타입:", type(arr))
```

## 2. 배열의 기본 속성 확인

NumPy 배열은 모양, 차원, 원소 개수, 자료형을 쉽게 확인할 수 있다.

```python
arr = np.array([1, 2, 3, 4, 5])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])

print("1차원 배열:", arr)
print("arr.shape:", arr.shape)
print("arr.ndim:", arr.ndim)
print("arr.size:", arr.size)
print("arr.dtype:", arr.dtype)

print()
print("2차원 배열:\n", arr2)
print("arr2.shape:", arr2.shape)
print("arr2.ndim:", arr2.ndim)
```

## 3. 인덱싱과 슬라이싱

배열의 특정 위치 값을 꺼내거나 일부 구간만 잘라낼 수 있다.

```python
arr = np.array([10, 20, 30, 40, 50])

print("arr[0] =", arr[0])
print("arr[2] =", arr[2])
print("arr[-1] =", arr[-1])
print("arr[1:4] =", arr[1:4])
```

## 4. 간단한 배열 연산

NumPy 배열은 같은 크기의 배열끼리 더하거나, 모든 원소에 같은 수를 더할 수 있다.

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("arr1 + arr2 =", arr1 + arr2)
print("arr1 * 2 =", arr1 * 2)
print("arr1 + 10 =", arr1 + 10)
```

## 5. 자주 쓰는 기초 함수

합계, 평균, 최댓값, 최솟값은 NumPy 함수로 쉽게 구할 수 있다.

```python
arr = np.array([1, 2, 3, 4, 5])

print(np.sum(arr))   # 합계
print(np.mean(arr))  # 평균
print(np.max(arr))   # 최댓값
print(np.min(arr))   # 최솟값
```

## 6. 배열 모양 바꾸기

`reshape()`를 사용하면 1차원 배열의 모양을 바꿀 수 있다.

```python
arr = np.array([1, 2, 3, 4, 5, 6])
arr2 = arr.reshape(2, 3)

print("원래 shape:", arr.shape)
print("바꾼 shape:", arr2.shape)
print(arr2)
```
