# Week 07 Matplotlib 히스토그램 실습

이번 실습은 `정렬 -> 요약 통계 -> 히스토그램` 흐름으로 진행합니다.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 0. 샘플 데이터 만들기

```python
np.random.seed(42)

scores = np.random.normal(loc=70, scale=12, size=200)
group_a = np.random.normal(loc=65, scale=10, size=200)
group_b = np.random.normal(loc=75, scale=14, size=200)

df_scores = pd.DataFrame({
    "score": scores,
    "group": np.where(scores >= np.mean(scores), "high_side", "low_side")
})

df_scores.head()
```

## 1. `np.sort()`로 작은 값부터 큰 값까지 정렬하기

```python
sorted_scores = np.sort(scores)

print("앞 10개:", np.round(sorted_scores[:10], 2))
print("뒤 10개:", np.round(sorted_scores[-10:], 2))
```

## 2. 범위, 중심, 퍼짐을 먼저 보기

```python
score_min = np.min(scores)
score_max = np.max(scores)
score_mean = np.mean(scores)
score_median = np.median(scores)
score_std = np.std(scores)
q1, q2, q3 = np.percentile(scores, [25, 50, 75])

print(f"min: {score_min:.2f}")
print(f"max: {score_max:.2f}")
print(f"mean: {score_mean:.2f}")
print(f"median: {score_median:.2f}")
print(f"std: {score_std:.2f}")
print(f"Q1: {q1:.2f}, Q2: {q2:.2f}, Q3: {q3:.2f}")
```

## 3. 가장 기본적인 히스토그램

```python
plt.figure(figsize=(7, 4))
plt.hist(scores)
plt.title('Basic Histogram')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

## 4. `bins`를 직접 지정하기

```python
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=12, color='skyblue', edgecolor='black')
plt.title('Histogram with bins=12')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

## 5. `bins` 개수 차이 비교

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

for ax, bin_count in zip(axes, [5, 12, 30]):
    ax.hist(scores, bins=bin_count, color='lightcoral', edgecolor='black')
    ax.set_title(f'bins={bin_count}')
    ax.set_xlabel('score')
    ax.set_ylabel('count')

plt.tight_layout()
plt.show()
```

## 6. 색상, 경계선, 투명도

```python
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=15, color='seagreen', edgecolor='white', alpha=0.8)
plt.title('Color, Edge, Alpha')
plt.xlabel('score')
plt.ylabel('count')
plt.grid(axis='y', alpha=0.25)
plt.show()
```

## 7. 두 분포를 겹쳐서 비교하기

```python
plt.figure(figsize=(8, 4))
plt.hist(group_a, bins=15, alpha=0.5, label='group_a', color='royalblue')
plt.hist(group_b, bins=15, alpha=0.5, label='group_b', color='orange')
plt.title('Overlayed Histograms')
plt.xlabel('score')
plt.ylabel('count')
plt.legend()
plt.show()
```

## 8. `density=True`로 보기

```python
plt.figure(figsize=(8, 4))
plt.hist(group_a, bins=15, alpha=0.5, density=True, label='group_a', color='royalblue')
plt.hist(group_b, bins=15, alpha=0.5, density=True, label='group_b', color='orange')
plt.title('Density Histogram')
plt.xlabel('score')
plt.ylabel('density')
plt.legend()
plt.show()
```

## 9. `cumulative=True`로 누적 히스토그램

```python
plt.figure(figsize=(7, 4))
plt.hist(scores, bins=20, cumulative=True, color='mediumpurple', edgecolor='black')
plt.title('Cumulative Histogram')
plt.xlabel('score')
plt.ylabel('cumulative count')
plt.show()
```

## 10. `ax.hist()`로 축 객체 기준으로 그리기

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.hist(scores, bins=15, color='steelblue', edgecolor='black')
ax.set_title('Histogram drawn with ax.hist()')
ax.set_xlabel('score')
ax.set_ylabel('count')
plt.show()
```

## 11. `Pandas` 열과 `sort_values()`

```python
sorted_series = df_scores['score'].sort_values()

print('오름차순 앞 5개')
print(np.round(sorted_series.head(5), 2))
print('\n오름차순 뒤 5개')
print(np.round(sorted_series.tail(5), 2))
```

## 12. `Pandas` 열을 바로 히스토그램에 넣기

```python
plt.figure(figsize=(7, 4))
plt.hist(df_scores['score'], bins=15, color='tomato', edgecolor='black')
plt.title('Histogram from a Pandas Series')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

## 13. 히스토그램과 막대그래프 비교 메모

- `hist()`는 연속형 수치 데이터를 구간별로 자동 집계합니다.
- `bar()`는 이미 집계된 범주별 값을 표현합니다.
