```python
import numpy as np
import sys

list_data = [1,2,3,4,5]
print("List 메모리 크기:", sys.getsizeof(list_data) * len(list_data))

array_data = np.array([1,2,3,4,5])
print("NumPy 배열 메모리 크기:", array_data.nbytes)
```

```python
import time

size = 10**6

list1 = list(range(size))
list2 = list(range(size))

start = time.time()
result_list = [x + y for x, y in zip(list1,list2)]
end = time.time()
print("리스트 연산 시간:", end - start)

arr1 = np.array(list1)
arr2 = np.array(list2)

start = time.time()
result_array = arr1 + arr2
end = time.time()
print("NumPy 연산 시간:", end - start)
```

```python
arr = np.array([1, 2, 3, 4])

print("배열:", arr)
print("평균:", np.mean(arr))
print("합계:", np.sum(arr))
print("최댓값:", np.max(arr))
print("최솟값:", np.min(arr))
print("표준편차:", np.std(arr))
```
